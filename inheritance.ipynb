Single Inheritance is a type of inheritance in which one child class inherits the properties and methods of only one parent class.

üëâ Matlab:
Ek class (child class) sirf ek hi class (parent class) se data aur functions ko inherit karti hai.

Single Inheritance helps in code reusability, reducing duplication, easy maintenance, and better program organization.

| üî¢ No. | Type of Inheritance      | Definition                                                               | Example Structure     |

| 5Ô∏è‚É£    | Hybrid Inheritance       | Combination of two or more types of inheritance.                         | Combination Structure |
